```python
import twint
config = twint.Config()
config.Proxy_host = "51.158.68.68"
config.Proxy_port = "8761"
config.Proxy_type = "http"
```

```python
import time
c = twint.Config()
c.Search = 'Ridwan Kamil'
c.Pandas = True
c.Limit = 500
c.Store_csv = True
#c.Custom["username"] = ["username"]
c.Custom["tweet"] = ["tweet"]
#c.Custom["location"] = ["place"]
c.Output = "crawltwitter_ridwankamil.csv"
try:
  twint.run.Search(c)
except :
  time.sleep(10)
  twint.run.Search(c)
```

```python
def column_names():
  return twint.output.panda.Tweets_df.columns

def twint_to_pd(columns):
  return twint.output.panda.Tweets_df[columns]
```
