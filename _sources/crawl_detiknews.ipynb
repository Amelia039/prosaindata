```python
import requests as re 
from bs4 import BeautifulSoup as bs
from urllib.request import Request, urlopen
import csv
import pandas as pd
```

```python
header = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
url = 'https://www.detik.com/search/searchall?query=ridwan+kamil&siteid=2&sortby=time'

titles = []
contents = []

#proses perulangan untuk crawling data
for page in range(1,9):
  req = re.get(url+str(page), headers=header)
  soup = bs(req.text, 'html.parser')
  items = soup.findAll('article')
  for it in items:
    link = it.find('a')['href']
    title = it.find('h2', 'title').text 
    titles.append(title)

    if link != '':
      req2 = re.get(str(link), headers=header)
      soup2 = bs(req2.text, 'html.parser')
      items2 = soup2.findAll("div", "detail__body-text itp_bodycontent")
      for it2 in items2:
        isi = it2.find('p').text

        if isi != '':
          contents.append(isi)


JudulBerita = pd.DataFrame(titles, columns=["Judul Berita"])
IsiBerita = pd.DataFrame(contents, columns=["Isi Berita"])

data = pd.concat([JudulBerita, IsiBerita], axis=1)
data = data.loc[0:50]
data
```
