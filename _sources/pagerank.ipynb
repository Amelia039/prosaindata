## PAGERANK

### PENGERTIAN


PageRank adalah algoritma yang digunakan oleh mesin pencari Google untuk menentukan urutan halaman web dalam hasil pencarian. Algoritma ini mengevaluasi jumlah dan kualitas tautan yang mengarah ke sebuah halaman untuk menentukan seberapa penting halaman tersebut. PageRank juga dapat digunakan untuk menganalisis jaringan sosial, serta jaringan kepercayaan atau peringkat kepercayaan pada perangkat lunak seperti email. Algoritma PageRank menghitung nilai penting (ranking) dari setiap node pada suatu grafik berbobot (weighted graph) dengan merangkai nilai penting dari node lain yang terhubung dengannya secara langsung dan tidak langsung.

### RUMUS

$PR(A) = (1-d) + d \sum\limits_{i \in B_{A}} \frac{PR(i)}{L(i)}$

di mana:
- $PR(A)$ adalah PageRank dari halaman A
- $d$ adalah faktor damping yang biasanya diatur ke 0,85
- $B_{A}$ adalah himpunan semua halaman yang memiliki tautan ke halaman A
- $L(i)$ adalah jumlah tautan keluar dari halaman i
- $PR(i)$ adalah PageRank dari halaman i

### MEMBUAT FUNGSI readDoc

* Fungsi readDoc() bertujuan untuk membaca sebuah dokumen yang diinputkan pengguna melalui terminal, kemudian mengembalikan isi dari dokumen tersebut dalam bentuk string.

* Fungsi ini pertama-tama meminta pengguna untuk memasukkan nama file. Selanjutnya, fungsi ini akan menentukan jenis file dokumen yang dimasukkan pengguna, apakah berformat .txt, .pdf, atau bukan keduanya. Jika dokumen berformat .txt, maka fungsi akan membuka file tersebut dan membaca isi dokumen ke dalam variabel document. Jika dokumen berformat .pdf, maka fungsi akan membuka file tersebut dan membaca teks pada halaman pertama ke dalam variabel document. Jika dokumen bukan berformat .txt atau .pdf, maka fungsi akan mengembalikan string kosong.

* Fungsi ini menggunakan library PyPDF2 untuk membaca dokumen PDF, serta menggunakan built-in function open() pada Python untuk membaca dokumen teks.

```python
def readDoc():
    name = input('Please input a file name: ') 
    print('You have asked for the document {}'.format(name))

    # now read the type of document
    if name.lower().endswith('.txt'):
        choice = 1
    elif name.lower().endswith('.pdf'):
        choice = 2
    else:
        choice = 3
        # print(name)
    print(choice)
    # Case 1: if it is a .txt file
        
    if choice == 1:
        f = open(name, 'r')
        document = f.read()
        f.close()
            
    # Case 2: if it is a .pdf file
    elif choice == 2:
        pdfFileObj = open(name, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        pageObj = pdfReader.getPage(0)
        document = pageObj.extractText()
        pdfFileObj.close()
    
    # Case 3: none of the format
    else:
        print('Failed to load a valid file')
        print('Returning an empty string')
        document = ''
    
    print(type(document))
    return document
```

### MEMBUAT FUNGSI TOKENISASI

* Fungsi tokenize(document) digunakan untuk melakukan tokenisasi terhadap dokumen yang diberikan menjadi kalimat-kalimat terpisah.

* Fungsi ini menggunakan library PunktSentenceTokenizer dari nltk.tokenize.punkt untuk melakukan tokenisasi. PunktSentenceTokenizer adalah sebuah kelas yang digunakan untuk membagi sebuah dokumen menjadi kalimat-kalimat berdasarkan aturan pemisahan yang sudah ditentukan sebelumnya.

* Fungsi tokenize(document) menerima sebuah dokumen (dalam bentuk string) sebagai input, kemudian menggunakan PunktSentenceTokenizer untuk membagi dokumen tersebut menjadi kalimat-kalimat terpisah. Hasilnya, fungsi ini mengembalikan daftar (list) yang berisi setiap kalimat dari dokumen sebagai elemen-elemen dalam daftar tersebut.

```python
def tokenize(document):
    # We are tokenizing using the PunktSentenceTokenizer
    # we call an instance of this class as sentence_tokenizer
    doc_tokenizer = PunktSentenceTokenizer()
    
    # tokenize() method: takes our document as input and returns a list of all the sentences in the document
    
    # sentences is a list containing each sentence of the document as an element
    sentences_list = doc_tokenizer.tokenize(document)
    return sentences_list
```

### PERHITUNGAN PAGERANK

dilakukan perhitungan nilai PageRank menggunakan fungsi nx.pagerank() dari pustaka NetworkX. PageRank digunakan untuk memberi nilai pada setiap kalimat dalam dokumen berdasarkan kesamaannya dengan kalimat lain dalam dokumen tersebut. Semakin sering sebuah kalimat muncul dalam hubungan yang erat dengan kalimat lainnya, semakin tinggi nilai PageRank yang akan diterimanya.

Variabel ranks adalah sebuah kamus (dictionary) yang memiliki key berupa kalimat dalam dokumen dan value berupa nilai PageRank dari kalimat tersebut. Dalam loop for, setiap kalimat dan nilai PageRank nya dicetak ke layar. Kemudian dilakukan analisis tipe data dari variabel ranks dan ukuran memorinya dalam bytes.

```python
# ranks is a dictionary with key=node(sentences) and value=textrank (the rank of each of the sentences)
ranks = nx.pagerank(nx_graph)

# analyse the data type of ranks
print(type(ranks))
print('The size used by the dictionary in Bytes is: {}'.format(sys.getsizeof(ranks)))

# print the dictionary
for i in ranks:
    print(i, ranks[i])
```

```python
# enumerate method: returns an enumerate object
# Use of list Comprehensions
# O/p: sentence_array is the sorted(descending order w.r.t. score value) 2-d array of ranks[sentence] and sentence 
# For example, if there are two sentences: S1 (with a score of S1 = s1) and S2 with score s2, with s2>s1
# then sentence_array is [[s2, S2], [s1, S1]]
sentence_array = sorted(((ranks[i], s) for i, s in enumerate(sentences_list)), reverse=True)
sentence_array = np.asarray(sentence_array)
```

```python
# as sentence_array is in descending order wrt score value
# fmax is the largest score value(the score of first element)
# fmin is the smallest score value(the score of last element)

rank_max = float(sentence_array[0][0])
rank_min = float(sentence_array[len(sentence_array) - 1][0])
```

```python
# print the largest and smallest value of scores of the sentence
print(rank_max)
print(rank_min)
```

```python
# Normalization of the scores
# so that it comes out in the range 0-1
# fmax becomes 1
# fmin becomes 0
# store the normalized values in the list temp_array

temp_array = []

# if all sentences have equal ranks, means they are all the same
# taking any sentence will give the summary, say the first sentence
flag = 0
if rank_max - rank_min == 0:
    temp_array.append(0)
    flag = 1

# If the sentence has different ranks
if flag != 1:
    for i in range(0, len(sentence_array)):
        temp_array.append((float(sentence_array[i][0]) - rank_min) / (rank_max - rank_min))

print(len(temp_array))
```

```python
# Calculation of threshold:
# We take the mean value of normalized scores
# any sentence with the normalized score 0.2 more than the mean value is considered to be 
threshold = (sum(temp_array) / len(temp_array)) + 0.2
```

```python
# Separate out the sentences that satiasfy the criteria of having a score above the threshold
sentence_list = []
if len(temp_array) > 1:
    for i in range(0, len(temp_array)):
        if temp_array[i] > threshold:
                sentence_list.append(sentence_array[i][1])
else:
    sentence_list.append(sentence_array[0][1])
```
